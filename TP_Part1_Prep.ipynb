```python
# Import necessary librries.
import io
import matplotlib
import matplotlib.pyplot as plt
import numpy
import pandas as pd
import seaborn
```

```python
#Lectura del dataset.
url = 'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/sysarmy_survey_2020_processed.csv'
df = pd.read_csv(url)
# Crea un índice desde 1 al total para trabajar más cómodamente. 
df = df.assign(id = df.index+1)
```

```python
# Trata los lenguajes para depurarlos en una sola columna. 

##Crea un nuevo df sólo con los lenguajes. 1 columna. 
prog = df[['id','tools_programming_languages']]
prog = prog.rename(columns={'tools_programming_languages':'lang'})
prog = prog.dropna()
# Just one NaN record in languages.

# Ej Remplaza la ocurrencia de 'xsl y xml.' con 'xsl,xml' -> replace('xsl y xml.','xsl,xml')
prog = prog.assign(lang = prog.lang.str.replace('VBA - Visual Basic Script','VBA,Visual Basic Script'))
prog = prog.assign(lang = prog.lang.str.replace('xsl y xml.','xsl,xml'))
prog = prog.assign(lang = prog.lang.str.replace('Power shell','PowerShell',regex=False))
prog = prog.assign(lang = prog.lang.str.replace('Power Shell','PowerShell',regex=False))
prog = prog.assign(lang = prog.lang.str.replace('Power Builder','PowerBuilder',regex=False))
prog = prog.assign(lang = prog.lang.str.replace('Oracle PL/SQL','PL/SQL',regex=False))
prog = prog.assign(lang = prog.lang.str.replace('Plsql','PL/SQL',regex=False))
prog = prog.assign(lang = prog.lang.str.replace('plsql','PL/SQL',regex=False))
prog = prog.assign(lang = prog.lang.str.replace('Pl-sql','PL/SQL',regex=False))
prog = prog.assign(lang = prog.lang.str.replace('Lenguaje Propio de PeopleSoft',',PeopleSoft'))
prog = prog.assign(lang = prog.lang.str.replace(' pero RPA',',RPA'))
prog = prog.assign(lang = prog.lang.str.replace('JSON y Powershell','JSON,Powershell'))
prog = prog.assign(lang = prog.lang.str.replace('HTML y CSS','HTML,CSS,'))
prog = prog.assign(lang = prog.lang.str.replace('Apex / salesforce','Apex salesforce'))
prog = prog.assign(lang = prog.lang.str.replace(' / ',','))
prog = prog.assign(lang = prog.lang.str.replace(' | ',',',regex=False))
```

```python
#Warning inservible. Ignorar. 
```

```python
# Genera una tabla de lenguajes con columnas separadas por 
# cada lenguaje. Crea una matriz de personas lenguajes.
# las personas se identifican con el id.
prog2 = pd.merge(prog,
                prog.lang.str.split(',', expand=True), ## Expand = True es lo que traspone la df.
                left_index=True, right_index=True)
prog2 = prog2.drop(columns=['lang'])
```

```python
prog2
```

```python
# Construye la lista vertical de id-persona / lenguaje
progLong = prog2.melt(id_vars=['id']).sort_values(['id'])
```

```python
progLong['value']
a = progLong[progLong.value=='Ninguno']
a
#seguimos limpiando. 
```

```python
progLong = progLong.dropna()
```

```python
progLong = progLong.assign(LANGUAGE=progLong.value.str.strip())
progLong
```

```python
#drop variable and value.
progLong = progLong.drop(columns=['value','variable'])
```

```python
#todo a minúscula.
progLong = progLong.assign(lang=progLong.LANGUAGE.str.lower())
progLong
```

```python
# Elimina lenguajes inválidos:
progLong = progLong[progLong.lang!='']
progLong = progLong[progLong.lang!='no']
progLong = progLong[~progLong.lang.str.contains('ingun|no apl|no es|no programo|no son')]
progLong = progLong[~progLong.lang.str.contains('laburo pedorro')]

progLong = progLong[(progLong.lang!='.') & (progLong.lang!='-')]
progLong = progLong.sort_values('lang')
```

```python
progLong
```

```python
#guarda la frecuencia por lenguaje.
x = progLong.groupby(['lang','LANGUAGE']).size().reset_index()
```

```python
x
```

```python
#Elimina duplicados
progLong = progLong[['id','lang']].drop_duplicates()
```

```python
#df final de lenguajes id de persona y lenguajes.
progLong
```

```python
#Ahora creo un df sólo con salarios.
# crea un df para salarios y renombra las columnas.
sal = df[['id','salary_monthly_BRUTO','salary_monthly_NETO']]
sal = sal.rename(columns={'salary_monthly_BRUTO':'sal_gross',
                          'salary_monthly_NETO':'sal_net'})
```

```python
sal
```

```python
# Ahora junto las dos df lang y sal.
#merge progams and salaries
progSalLong = pd.merge(progLong,sal, on='id')
# Agrego una columna que cuenta la frecuencia. 
```

```python
progSalLong
```

```python
#Agrego frecuencia, promedio, mediana y desviación.
progSalLong['langN'] = progSalLong.groupby('lang')['id'].transform('count')
progSalLong['MeanSalGross'] = progSalLong.groupby('lang')['sal_gross'].transform('median')
progSalLong['MedianSalGross'] = progSalLong.groupby('lang')['sal_gross'].transform('mean')
progSalLong['DevSalGross'] = progSalLong.groupby('lang')['sal_gross'].transform('std')
```
