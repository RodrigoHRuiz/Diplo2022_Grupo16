PASO 8: Como se anticipó en el paso 5, previo a aplicar la reducción de dimensionalidad, sería necesario escalar los datos para evitar que alguna variable tome mayor relevancia sobre las otras.<br>
Se toma el datafram del punto anterior, donde los datos ya están escalados y se aplica el método de descomposición PCA de scikit-learn.

```python
airbnb_melb_all.sample(10)
```

```python
feature_cols = ['precio', 'habitaciones', 'baños', 'cp_melb', 'año_construccion',
            'sup_construida', 'media_precio_Airbnb', 'tipo_h', 'tipo_t', 'tipo_u', 'region_Northern Metropolitan', 'region_Northern Victoria', 'region_South-Eastern Metropolitan', 'region_Southern Metropolitan', 'region_Western Metropolitan']
feature_dict = list(airbnb_melb_all[feature_cols].T.to_dict().values())
feature_dict[:2]
```

```python
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer()
feature_matrix = vec.fit_transform(feature_dict)
```

```python
feature_matrix
```

```python
vec.get_feature_names()
```

```python
if np.any(np.isnan(feature_matrix.data)):
  feature_matrix.data = np.nan_to_num(feature_matrix.data)
```

```python
# Es necesario calcular el tamaño de la matriz resultante
matrix_size_mb = feature_matrix.shape[0] * feature_matrix.shape[1] * 4 / 1024 / 1024
print("La matriz densa pesaria aproximadamente {:.2f} MB".format(matrix_size_mb))

limit_size_mb = 10
precision_type = np.float32
if matrix_size_mb < limit_size_mb:  # Matrix is less than 10MB
  dense_feature_matrix = feature_matrix.astype(precision_type).todense()
else:
  # Se calcula cuántas filas se toman, dado el número de columnas de la matriz
  n_rows = int(limit_size_mb *1024 * 1024 / 4 / feature_matrix.shape[1])
  print("Matriz demasiado grande! Usar solo las primeras {} de {} filas".format(
      n_rows, feature_matrix.shape[0]))
  dense_feature_matrix = feature_matrix[:n_rows].astype(precision_type).todense()

print("Tamaño final: {:.2f}".format(dense_feature_matrix.nbytes / 1024 / 1024))
```

<font color='Green'><strong>
Se incluyen todas las filas de datos en la matriz, ya que el tamaño resultante está por debajo del límite marcado.

```python
if dense_feature_matrix.shape[0] < len(airbnb_melb_all):
 airbnb_melb_mice_sample = airbnb_melb_all.iloc[:dense_feature_matrix.shape[0]]
else:
  airbnb_melb_mice_sample = airbnb_melb_all
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
pca.fit(dense_feature_matrix)
proyected_features = pca.transform(dense_feature_matrix)
proyected_features.shape
```

```python
print("2 Principal components")
print(pca.components_[:3])
print("Explained variance ratio")
print(pca.explained_variance_ratio_)
```

<font color='Green'><strong>
Se puede ver que con las dos primeras componentes se explica el 68% de la varianza, y esto asciende al 80% si se considera una tercera componente.
Se grafican los puntos, según las dos primeras componentes.

```python
airbnb_melb_mice_sample[['pca1_scaled', 'pca2_scaled']] = proyected_features[:,:2]
```

```python
sns.boxplot(airbnb_melb_mice_sample.precio)
```

```python
data = airbnb_melb_mice_sample.sample(2000).sort_values(by=['tipo_h','tipo_t','tipo_u'],ascending=False, inplace=False)
fig = plt.figure(figsize=(10, 10))
hue_serie = data['tipo_h'].astype(str) + ', ' + data['tipo_t'].astype(str) + ', ' + data['tipo_u'].astype(str)
g = sns.scatterplot(data=data, x='pca1_scaled', y='pca2_scaled',\
                    hue=hue_serie,legend=True)
h, l = g.get_legend_handles_labels()
plt.legend(title='tipo', handles=h, labels=['house', 'townhouse', 'unit'])
```

Tenemos que volver a escalar todos los datos a un mismo rango. ¿Utilizamos estandarización o escalado linear?


```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
stdandard_dense_feature_matrix = std_scaler.fit_transform(dense_feature_matrix)
np.max(stdandard_dense_feature_matrix), np.min(stdandard_dense_feature_matrix)
```

```python
from sklearn.decomposition import PCA
pca_std = PCA(n_components=15)
pca_std.fit(stdandard_dense_feature_matrix)
proyected_features_std = pca_std.transform(stdandard_dense_feature_matrix)
proyected_features_std.shape
```

```python
print("2 Principal components")
print(pca_std.components_[:3])
print("Explained variance ratio")
print(pca_std.explained_variance_ratio_)
```

<font color='Green'><strong>
En este caso, se puede ver que con las dos primeras componentes se explica el 41% de la varianza, y esto asciende al 55% si se considera una tercera componente.<br>
Se trazan en un mismo gráfico la proporción de la varianza explicada según el escalado o estandarizado.<br>
Como se podrá apreciar, con el escalado se logra explicar mayor proporción de varianza en las dos o tres primeras componentes, frente al estandarizado de los datos.

```python
fig = plt.figure(figsize=(10, 5))
sns.lineplot(x=range(pca_std.n_components),
                 y=pca_std.explained_variance_ratio_, label='Estandarizado')
sns.lineplot(x=range(pca.n_components),
                 y=pca.explained_variance_ratio_, label='Escalado')
plt.ylabel("Proporción de la varianza explicada")
plt.xlabel("n Componentes")
plt.legend()
```
